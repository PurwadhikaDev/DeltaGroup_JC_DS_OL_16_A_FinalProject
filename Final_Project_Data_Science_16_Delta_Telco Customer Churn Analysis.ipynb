# Business Problem Understanding

## Context

Pelanggan selalu menjadi teka-teki bagi sebuah bisnis. Ada banyak sekali buku tentang cara meraih kesuksesan dalam bisnis, tetapi jika ditelaah lebih dalam, semuanya tetap kembali pada pelanggan karena merekalah yang membawa keuntungan.

Sebuah bisnis, termasuk Telco sebagai perusahaan yang berfokus pada internet dan telekomunikasi, tentu ingin selalu memberikan pelayanan terbaik kepada pelanggan agar mereka tetap dapat bertahan di tengah persaingan bisnis yang kompetitif. Namun, ungkapan *"Nothing Last Forever"* selalu relevan dalam kehidupan sehari-hari, termasuk dalam hubungan antara pelangganan dan langganan. Oleh karena itu, memahami pola perilaku pelanggan menjadi sangat penting untuk menjaga stabilitas bisnis serta mendukung pengambilan keputusan terkait retensi dan loyalitas pelanggan. Di sinilah peran analisis Customer Churn menjadi krusial.

*Customer Churn* adalah tingkat di mana pelanggan meninggalkan bisnis atau platform komersial dan mengalihkan pengeluarannyaa tempat lain. *Customer Churn* menjadi metrik yang penting karena memperoleh pelanggan baru bisa lima hingga tujuh kali lebih mahal dibandingkan mempertahankan pelanggan yang sudah ada<sup>[1](https://www.forbes.com/councils/forbesbusinesscouncil/2022/12/12/customer-retention-versus-customer-acquisition/), [2](https://www.outboundengine.com/blog/customer-retention-marketing-vs-customer-acquisition-marketing/#:~:text=Customer%20acquisition%20%26%20retention%20marketing%20stats,profits%20from%2025%2D95%25.)</sup>. Untuk mengurangi tingkat *churn*, perusahaan perlu memprediksi pelanggan mana yang berisiko tinggi untuk berhenti berlangganan.

Laporan ini mengeksplorasi lebih dalam pola perilaku pelanggan pascabayar di perusahaan Telco guna mengungkap wawasan dan strategi dalam meningkatkan retensi serta kepuasan pelanggan dalam industri telekomunikasi, sehingga dapat meminimalkan risiko *churn*.

## Problem Statement

Pelanggan dapat dengan mudah berganti paket bahkan berganti provider dalam industri telekomunikasi. Menurut laporan dari [MaxBill (Maret, 2023)](https://www.dataportabilitycooperation.org/data-portability-telecom-customer-churn-rates/#:~:text=As%20reported%20by%20MaxBill%20in%20March%202023%2C%20the,times%20more%20costly%20than%20retaining%20an%20existing%20one), industri ini menghadapi tingkat *churn rate* sebesar 20-40 persen, yang menunjukkan peningkatan signifikan dalam perpindahan pelanggan.

Telco, sebagai perusahaan telekomunikasi, terus dihadapkan pada tantangan dalam menentukan langkah terbaik untuk mengatasi tingkat *churn* yang cukup signifikan, yaitu sebesar 26,5 persen. Untuk **menurunkan persentase *churn*, perusahaan perlu mengidentifikasi pelanggan yang memiliki potensi berhenti berlangganan**, sehingga dapat memutuskan strategi retensi yang lebih efisien.

*Stakeholder* dalam perusahaan telekomunikasi dapat menerapkan berbagai strategi retensi, seperti memberikan potongan harga, menawarkan paket layanan yang menarik, atau memberikan prioritas pelayanan, untuk mempertahankan pelanggan. Namun, kebijakan pemberian insentif retensi belum sepenuhnya efektif. Apabila tidak ditargetkan dengan tepat, insentif ini dapat meningkatkan biaya operasional dan mengurangi potensi keuntungan, terutama jika diberikan kepada pelanggan yang sebenarnya sudah loyal dan tidak berencana untuk berhenti berlangganan.

4. Pada feature `TotalCharges`, pelanggan yang *churn* cenderung memiliki total tagihan yang lebih rendah. Hal ini konsisten dengan pola distribusi `tenure`, di mana pelanggan yang berhenti berlangganan belum membayar akumulasi tagihan dalam jumlah besar.

```python
def plot_distribution(df, column_name):
    # Pemetaan berdasarkan kolom yang dipilih
    mappings = {
        'gender': {'Male': 'Male', 'Female': 'Female'},
        'SeniorCitizen': {0: 'No', 1: 'Yes'},
        'Partner': {'No': 'No', 'Yes': 'Yes'},
        'Dependents': {'No': 'No', 'Yes': 'Yes'},
        'PhoneService': {'No': 'No', 'Yes': 'Yes'},
        'MultipleLines': {'No': 'No', 'Yes': 'Yes', 'No phone service': 'No phone service'},
        'InternetService': {'DSL': 'DSL', 'Fiber optic': 'Fiber Optic', 'No': 'No'},
        'OnlineSecurity': {'No': 'No', 'Yes': 'Yes', 'No internet service': 'No internet service'},
        'OnlineBackup': {'No': 'No', 'Yes': 'Yes', 'No internet service': 'No internet service'},
        'DeviceProtection': {'No': 'No', 'Yes': 'Yes', 'No internet service':'No internet service'},
        'TechSupport': {'No': 'No', 'Yes': 'Yes', 'No internet service': 'No internet service'},
        'StreamingTV': {'No': 'No', 'Yes': 'Yes', 'No internet service': 'No internet service'},
        'StreamingMovies': {'No': 'No', 'Yes': 'Yes', 'No internet service': 'No internet service'},
        'Contract': {'Month-to-month': 'Month-to-Month', 'One year': 'One Year', 'Two year': 'Two-Year'},
        'PaperlessBilling': {'No': 'No', 'Yes': 'Yes'}
    }

     # Menghitung jumlah kategori berdasarkan kolom yang dipilih
    column_counts = df[column_name].value_counts()

    # Menerapkan pemetaan ke indeks jika kolom ada dalam pemetaan
    if column_name in mappings:
        column_counts.index = column_counts.index.map(mappings[column_name])

    # Menghitung distribusi berdasarkan churn
    churn_counts = pd.crosstab(df[column_name], df['Churn'])

    # Menerapkan pemetaan pada indeks churn_counts jika kolom ada dalam pemetaan
    if column_name in mappings:
        churn_counts.index = churn_counts.index.map(mappings[column_name])

    # Membuat subplot dengan dua grafik
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))

    # Menentukan warna untuk pie chart dan bar chart
    colors = ['#66b3ff', '#ff9999', '#99ff99', '#ffcc66', '#ff6666', '#cc66ff']  # Tambahkan warna baru jika lebih dari 2 kategori

    # Plot pertama: Pie Chart distribusi kolom yang dipilih
    axes[0].pie(column_counts, labels=column_counts.index, autopct='%1.1f%%', startangle=90, colors=colors[:len(column_counts)])
    axes[0].set_title(f'Distribusi {column_name.capitalize()}')

    # Plot kedua: Bar chart full-stack distribusi kolom berdasarkan churn
    churn_counts.plot(kind='bar', stacked=True, ax=axes[1], color=colors[:len(churn_counts.columns)])
    axes[1].set_title(f'Distribusi {column_name.capitalize()} Berdasarkan Churn')
    axes[1].set_ylabel('Jumlah')
    axes[1].set_xlabel(column_name.capitalize())

    # Menambahkan persentase pada barplot
    for p in axes[1].patches:
        height = p.get_height()
        width = p.get_width()
        x = p.get_x() + width / 2
        y = p.get_y() + height / 2
        total = churn_counts.sum().sum()  # total count untuk persentase
        percentage = (height / total) * 100
        axes[1].text(x, y, f'{percentage:.1f}%', ha='center', va='center', color='white')

    # Menampilkan plot
    plt.tight_layout()
    plt.show()
```

Oleh karena itu, analisis lebih lanjut dilakukan untuk melihat apakah terdapat keterkaitan antara fitur `MultipleLines` dan `PhoneService`.

```python
# Membuat tabel kontingensi untuk analisis pengaruh terhadap Churn
contingency_table = pd.crosstab([df['PhoneService'], df['MultipleLines']], df['Churn'])
contingency_table
```

```python
# Hitung persentase Churn untuk setiap kombinasi
contingency_table_percent = contingency_table.div(contingency_table.sum(axis=1), axis=0) * 100

# Plot heatmap
plt.figure(figsize=(10, 7))
sns.heatmap(contingency_table_percent, annot=True, cmap="YlGnBu", fmt=".2f", cbar_kws={'label': 'Churn Percentage'})
plt.title("Heatmap: Pengaruh PhoneService dan MultipleLines terhadap Churn")
plt.xlabel("Churn")
plt.ylabel("PhoneService dan MultipleLines")
plt.show()
```

Berdasarkan hasil analisis, terlihat bahwa:
1. Pelanggan dengan layanan `PhoneService` dan tambahan `MultipleLines` memiliki risiko *churn* sebesar 28,61 persen.
2. Pelanggan yang hanya memiliki `PhoneService` tanpa `MultipleLines` memiliki tingkat *churn* sebesar 25,04 persen.
3. Pelanggan yang tidak menggunakan kedua layanan tersebut memiliki tingkat *churn* sebesar 24,93 persen.

Namun, jumlah riil antara pelanggan *churn* yang menggunakan kedua layanan (850) dengan pelanggan *churn* yang hanya menggunakan layanan `PhoneService` (849) relatif sama. Dengan demikian, dapat disimpulkan bahwa pengaruh `PhoneService` dan `MultipleLines` terhadap tingkat *churn* tidak signifikan, sesuai dengan hipotesis awal.

### Internet Support Sebagai Bundling Layanan Tambahan

```python
df_internetSupport = df[["OnlineSecurity", "OnlineBackup", "DeviceProtection", "TechSupport", "Churn"]]
df_internetSupport["total_internetSupport"] = df_internetSupport[["OnlineSecurity", "OnlineBackup", "DeviceProtection", "TechSupport"]].apply(lambda x: (x == "Yes").sum(), axis=1)

internetSupport_data = df_internetSupport.groupby(['total_internetSupport', 'Churn']).size().unstack(fill_value=0)
internetSupport_percent = internetSupport_data.div(internetSupport_data.sum(axis=1), axis=0) * 100

ax = internetSupport_data.plot(kind='bar', stacked=True, figsize=(8, 6))

for i, bars in enumerate(ax.containers):
    total_counts = internetSupport_data.sum(axis=1)
    for bar, total in zip(bars, total_counts):
        height = bar.get_height()
        if height > 0:
            percentage = height / total * 100
            
            if percentage < 10:  
                ax.text(bar.get_x() + bar.get_width() / 2, bar.get_y() + height + 1, 
                        f'{percentage:.1f}%', ha='center', va='bottom', fontsize=9, color='black')
            else:
                ax.text(bar.get_x() + bar.get_width() / 2, bar.get_y() + height / 2, 
                        f'{percentage:.1f}%', ha='center', va='center', fontsize=10, color='white')

plt.title("Churn Berdasarkan Total Layanan Internet Support")
plt.xticks(rotation=0)
plt.show()
```

### Encoding

Proses ini mengubah data kategorikal menjadi numerik agar dapat digunakan dalam model karena hanya bisa dijalankan dengan angka.

Kami mengasumsikan bahwa "No Internet Service" dan "No Phone Service" memiliki arti yang sama dengan "No". Oleh karena itu, seluruh data dengan nilai tersebut akan diganti menjadi "No" di setiap kolom terkait. Hal ini bertujuan untuk menyederhanakan analisis dan memastikan konsistensi interpretasi.

```python
ganti = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV','StreamingMovies']
for i in ganti:
    df[i] = df[i].replace({'No internet service': 'No'})
```

```python
df['MultipleLines'] = df['MultipleLines'].replace({'No phone service': 'No'})
```

```python
listItem = []
for col in df.columns:
    unique_values = df[col].drop_duplicates()
    sample_size = min(3, len(unique_values))  
    listItem.append([col, df[col].dtype, df[col].isna().sum(), round((df[col].isna().sum()/len(df[col])) * 100, 2),
                    df[col].nunique(), list(unique_values.sample(sample_size).values)])

dfDesc = pd.DataFrame(columns=['dataFeatures', 'dataType', 'null', 'nullPct', 'unique', 'uniqueSample'],
                     data=listItem)
dfDesc
```

Data dengan isi "No Internet Service" dan "No Phone Service" sudah berubah menjadi "No".

Selanjutnya dilakukan ordinal mapping, di mana data kategorikal ordinal diubah menjadi angka berdasarkan urutan atau tingkatan.

```python
# ordinal mapping InternetService && Contract
internet_mapping = {
    "No": 0,
    "DSL": 1,
    "Fiber optic": 2
}
contract_mapping = {
    "Month-to-month": 0,
    "One year": 1,
    "Two year": 2
}

# transform
df["InternetService"] = df["InternetService"].map(internet_mapping)
df["Contract"] = df["Contract"].map(contract_mapping)
```

Selanjutnya dilakukan one-hot encoding untuk mengubah data kategorikal nominal menjadi bentuk numerik dengan merepresentasikan setiap kategori sebagai biner.

```python
# one hot PaymentMethod
payment_encoded = pd.get_dummies(df["PaymentMethod"], prefix="payment", drop_first=True).astype(int)
payment_encoded.columns = payment_encoded.columns.str.replace(" ", "_").str.lower()

df = pd.concat([df, payment_encoded], axis=1)
df.drop(columns=["PaymentMethod"], inplace=True)
```

```python
# mengubah feature menjadi binary
## feature gender
df["gender"] = df["gender"].map({"Female": 0, "Male": 1})

## rest with a Yes No value
binary_mapping = {
    "No": 0,
    "Yes": 1
}
binary_feature = ["Partner", "Dependents", "PhoneService", "MultipleLines", "OnlineSecurity", "OnlineBackup", "DeviceProtection",
                  "TechSupport", "StreamingTV", "StreamingMovies", "PaperlessBilling", "Churn"]

for feature in binary_feature:
    df[feature] = df[feature].map(binary_mapping)
```

Sebelumnya fitur `SeniorCitizen` memiliki tipe data object. Oleh karena itu, perlu diubah menjadi integer.

```python
# convert feature SeniorCitizen ke int
df["SeniorCitizen"] = df["SeniorCitizen"].astype(int)
```

```python
# Model Selection
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import recall_score, classification_report
```

## Benchmark Model

Untuk evaluasi model awal, kami akan menilai **6** jenis model yang berbeda:

1. Regresi Logistik [[1]](https://www.geeksforgeeks.org/understanding-logistic-regression/): Metode statistik yang digunakan untuk memprediksi hasil biner berdasarkan satu atau lebih variabel prediktor. Model ini menghitung probabilitas bahwa input tertentu termasuk dalam kategori tertentu, dan probabilitas ini kemudian diubah menjadi hasil biner (misalnya, jika probabilitas lebih besar dari 0,5, itu diklasifikasikan sebagai kategori 1; jika tidak, kategori 0). 

2. Support Vector Machine (SVM) [[2]](https://www.geeksforgeeks.org/support-vector-machine-algorithm/): salah satu algoritma pembelajaran mesin yang digunakan untuk tugas klasifikasi dan regresi. SVM bekerja dengan mencari hyperplane terbaik yang memisahkan data ke dalam dua kelas yang berbeda, dengan margin terbesar antara kedua kelas tersebut.

3. Decision Tree [[3]](https://scikit-learn.org/stable/modules/tree.html): Model yang membagi dataset menjadi subset berdasarkan nilai variabel input, dengan tujuan mengurangi metrik tertentu (seperti Gini impurity atau entropy) pada setiap pembagian. Struktur pohon dibangun dengan mengulang pembagian ini, dan setiap jalur dari akar hingga simpul daun mewakili serangkaian keputusan yang mengarah pada prediksi kelas akhir.

4. Random Forest [[4]](https://www.geeksforgeeks.org/random-forest-algorithm-in-machine-learning/): Metode ensemble yang membangun banyak pohon keputusan selama fase pelatihan. Pada tugas klasifikasi, model ini memberikan kelas yang paling sering diprediksi oleh pohon-pohon tersebut, sementara pada tugas regresi, model ini mengambil rata-rata prediksi dari semua pohon. Metode ini membantu mengurangi overfitting dan meningkatkan akurasi dengan merata-ratakan prediksi dari banyak pohon.

5. XGBoost (Extreme Gradient Boosting) [[5]](https://www.geeksforgeeks.org/ml-xgboost-extreme-gradient-boosting/): Metode ensemble yang membangun banyak pohon keputusan secara berurutan, di mana setiap pohon bertujuan untuk memperbaiki kesalahan dari pohon sebelumnya. Ini menggunakan teknik gradient boosting untuk meminimalkan fungsi kerugian melalui kombinasi prediksi pohon berturut-turut.

6. LightGBM (Light Gradient Boosting Machine) [[6]](https://www.geeksforgeeks.org/lightgbm-light-gradient-boosting-machine/): Kerangka kerja gradient boosting yang menggunakan algoritma pembelajaran berbasis pohon dan dioptimalkan untuk pelatihan yang terdistribusi dan efisien, terutama pada dataset besar.







### Training

```python
logreg = LogisticRegression()
svm = SVC()
dt = DecisionTreeClassifier()
rf = RandomForestClassifier()
xgb = XGBClassifier()
lgbm = LGBMClassifier()
```

```python
models = [logreg, svm, dt, rf, xgb, lgbm]
score_recall=[]
std=[]

for i in models:
    skfold=StratifiedKFold(n_splits=5)
    estimator=Pipeline([
        ('model',i)])
    model_cv=cross_val_score(estimator,X_train,y_train,cv=skfold,scoring='recall')
    score_recall.append(model_cv.mean())
    std.append(model_cv.std())
    
pd.DataFrame({
    'model':['Logistic Regression', 'SVM', 'Decision Tree', 'Random Forest', 'XGBoost', 'LightGBM'],
    'recall_score':score_recall,'sdev':std
}).set_index('model').sort_values(by='recall_score',ascending=False)
```

Training model ini merupakan pengujian suatu model pada data train, yang berguna untuk melihat model mana yang paling baik sebagai benchmark model. Dari hasil modeling didapat bahwa Logistic Regresion memiliki nilai recall `yang paling baik` dari 6 model yang diuji. Logistic Regression memiliki nilai 53% yang merupakan model terbaik dan Random Forest menjadi nilai dengan nilai recall terburuk dengan 48%.

Setelah melakukan modeling pada train data. Kami memutuskan untuk menggunakan 6 buah model dalam modeling dengan test data

### Test Data

```python
models = [logreg, svm, dt, rf, xgb, lgbm]
score_recall = []

def y_pred_func(i, X, y):
    x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

    estimator = Pipeline([
        ('model', i)
    ])

    estimator.fit(x_train, y_train)
    y_pred = estimator.predict(x_test)

    return estimator, y_pred, y_test

for i, j in zip(models, ['Logistic Regression', 'SVM', 'Decision Tree', 'Random Forest', 'XGBoost', 'LightGBM']):
    estimator, y_pred, y_test = y_pred_func(i, X_train, y_train)
    score_recall.append(recall_score(y_test, y_pred))
    
    print(j, '\n', classification_report(y_test, y_pred))

pd.DataFrame({
    'model': ['Logistic Regression', 'SVM', 'Decision Tree', 'Random Forest', 'XGBoost', 'LightGBM'],
    'recall_score': score_recall
}).set_index('model').sort_values(by='recall_score', ascending=False)
```

Dari tabel di atas, didapat bahwa **Logistic Regression masih menjadi model dengan nilai recall terbaik dengan nilai 53%**. Berdasarkan nilai recall tersebut, kami memutuskan menggunakan Logistic Regression sebagai model machine learning pada project ini, yang selanjutnya akan dilakukan resampling untuk menangani imbalance pada data

## Resampling: SMOTE & Penalized Model

Resampling adalah teknik yang digunakan dalam pembelajaran mesin untuk menangani masalah ketidakseimbangan kelas dalam dataset (imbalance). Ketidakseimbangan kelas sering terjadi ketika satu kelas lebih banyak jumlahnya dibandingkan dengan kelas lainnya dalam hal ini pada kolom churn, yang dapat menyebabkan model tidak dapat mempelajari pola dari kelas minoritas dengan baik. Dua pendekatan umum untuk mengatasi masalah ini adalah SMOTE (Synthetic Minority Over-sampling Technique) dan Penalized Model.

SMOTE bekerja dengan menciptakan data sintetis baru untuk kelas minoritas, membantu model mengenali pola yang lebih baik pada kelas yang lebih sedikit.

Penalized Model memberi penalti pada kesalahan klasifikasi untuk kelas minoritas, sehingga model lebih fokus pada kesalahan tersebut, tanpa memerlukan perubahan pada data.

```python
from imblearn.over_sampling import SMOTE
```

```python
# split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# SMOTE
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# penalized model
logreg_balanced = LogisticRegression(class_weight="balanced", random_state=42)
logreg_balanced.fit(X_train_resampled, y_train_resampled)

# predict
y_pred = logreg_balanced.predict(X_test)

print("Classification Report:\n", classification_report(y_test, y_pred))
print("Recall Score:", recall_score(y_test, y_pred))
```

Berdasarkan hasil recall yang didapat setelah melakukan sampling pada data, terjadi kenaikan pada nilai recall sebesar 21%, dengan hasil sebelum sampling sebesar 53%, dengan sampling smote dan penalized model naik menjadi 74%. Hal ini menunjukkan bahwa melakukan resampling pada data dapat meningkatkan kinerja model **(nilai recall)**.

## Feature Selection

Feature selection adalah proses pemilihan subset fitur (variabel) yang paling relevan dan penting untuk membangun model pembelajaran mesin yang efektif. Tujuan dari feature selection adalah untuk meningkatkan performa model, mengurangi kompleksitas, menghindari overfitting, dan mengurangi waktu komputasi.

```python
coef_df = pd.DataFrame({"Feature": X.columns, "Coefficient": logreg_balanced.coef_[0]})
coef_df["Abs_Coefficient"] = coef_df["Coefficient"].abs()

sorted_features = coef_df.sort_values(by="Abs_Coefficient", ascending=False)
sorted_features
```

Berdasarkan feature selection, dapat dilihat nilai absolute koeffisien yang paling tinggi adalah MonthlyCharges, dan yang paling rendah adalah gender. Hal ini menunjukkan bahwa kolom MonthlyCharges memiliki pengaruh paling besar dalam customer melakukan churn, sedangkan gender kurang memiliki pengaruh dalam keinginan customer untuk churn.

### Model Dengan 15 Feature Terpenting

Dikarenakan kami menggunakan Based Model, kami menggunakan feature dengan nilai absolute coeffisient diatas 0.45. Dengan melakukan sort terhadap nilai Abs_coefficient, didapat 15 Feature yang digunakan dalam modeling.

```python
selected_features = sorted_features["Feature"].head(15).tolist()

X_train_selectedFeatures = X_train_resampled[selected_features]
X_test_selectedFeatures = X_test[selected_features]

logreg_selectedFeatures = LogisticRegression(class_weight="balanced", random_state=42)
logreg_selectedFeatures.fit(X_train_selectedFeatures, y_train_resampled)

y_pred_selectedFeatures = logreg_selectedFeatures.predict(X_test_selectedFeatures)

print("Classification Report:\n", classification_report(y_test, y_pred_selectedFeatures))
print("Recall Score:", recall_score(y_test, y_pred_selectedFeatures))
```

Dengan menggunakan Logistic Regression dan menggunakan 15 feature yang telah dipilih di atas, kemudian dilakukan modeling ulang pada data test, didapat bahwa nilai recall yang dihasilkan sebesar 73%.

## Hyperparameter Tuning

Hyperparameter tuning adalah proses untuk menemukan kombinasi hyperparameter terbaik yang akan menghasilkan model pembelajaran mesin yang paling optimal. Hyperparameter adalah parameter yang ditetapkan sebelum proses pelatihan dimulai dan tidak dipelajari oleh model selama pelatihan. Model yang digunakan adalah Logistic Regression dengan parameter sebagai berikut:

1. `C`: [0.001, 0.01, 0.1, 1, 10, 100], 
2. `penalty`: ["l1", "l2", "elasticnet", "None"], 
3. `solver`: ["liblinear", "lbfgs", "newton-cg", "sag", "saga"], 
4. `max_iter`: [100, 200, 500]


Sumber: https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html

```python
from sklearn.model_selection import GridSearchCV
```

```python
hyperparam_space = {
    "C": [0.001, 0.01, 0.1, 1, 10, 100], # regularization strength
    "penalty": ["l1", "l2", "elasticnet", "None"], # type of regularization
    "solver": ["liblinear", "lbfgs", "newton-cg", "sag", "saga"], # algorithm
    "max_iter": [100, 200, 500]
}

logreg = LogisticRegression(class_weight="balanced", random_state=42)

grid_search = GridSearchCV(logreg, hyperparam_space, scoring="recall", cv=5, n_jobs=-1)
grid_search.fit(X_train_selectedFeatures, y_train_resampled)

best_model = grid_search.best_estimator_

y_pred_tuned = best_model.predict(X_test_selectedFeatures)

print("Best Parameters:", grid_search.best_params_)
print("Best Recall Score (mean):", grid_search.best_score_)
```
